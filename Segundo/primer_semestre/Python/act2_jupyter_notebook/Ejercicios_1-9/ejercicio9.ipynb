**Escriba un programa que solicite por teclado una palabra y calcule el valor de la misma dada la siguiente tabla de valores del juego Scrabble:**



Declaro las variables necesarias e inicio el contador

```python
word = input('Ingrese una palabra').upper()

one = ('A','E','I','O','U','L','N','R','S','T')
two = ('D','G')
three = ('B','C','M','P')
four = ('F','H','V','W','Y')
five = ('K')
eight = ('J','X')
ten = ('Q','Z')
value = 0

# se podria usar un diccionario
```

Analizo cada letra, sumo lo que corresponda e imprimo el resultado

```python
for letter in word:
    if letter in one: value += 1
    elif letter in two: value += 2
    elif letter in three: value += 3
    elif letter in four: value+= 4
    elif letter in five: value+= 5
    elif letter in eight: value+= 8
    else: value+= 10

print(f"El valor de {word} es {value}")
```
